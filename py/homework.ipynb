```python
#-*-coding:utf-8-*-
import numpy as np
import matplotlib.pyplot as plt

def func(x, c = 1):
    if x >= 2:
        return x**2
    else:
        return x**2 + c*(x - 2.0)**2

def plotFuncs():
    params = {1, 2, 10}
    color = ['r', 'g', 'b']
    xs = np.linspace(-1, 5, 50)
    for i, p in enumerate(params):
        ys = np.array([func(x, p) for x in xs])
        plt.plot(xs, ys, c = color[i])
        plt.scatter(xs, ys, c = color[i], s = 7)
    plt.grid(axis = 'both')
    plt.show()

plotFuncs()
```
